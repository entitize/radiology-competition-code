### Dumb baseline: predict the mean value for each category

```python
import pandas as pd
```

```python
train_df = pd.read_csv("../data/student_labels/train.csv")
```

```python
train_df.head()
```

```python
diseases = train_df.columns[5:]
disease_to_most_common = {}
for disease in diseases:
    disease_to_most_common[disease] = train_df[disease].mean()
print(disease_to_most_common)
```

```python
test_df = pd.read_csv("../data/student_labels/test_ids.csv")
test_df.head()
```

```python
for k, v in disease_to_most_common.items():
    test_df[k] = v
test_df.head()
```

```python
test_df.drop("Path", axis=1, inplace=True)
test_df.to_csv("../results/dumb_baseline.csv", index=False)
```
